```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
%matplotlib inline
import json
import math
```

```python
r = requests.get("http://www.the-numbers.com/movie/records/All-Time-Domestic-Box-Office")
soup = BeautifulSoup(r.text)
```

```python
allrow = soup.findAll('tr')
```

```python
el = allrow[1].findAll('td')
a = el[4].contents[-1]
for ch in ['$',',']:
    a = a.replace(ch,'')
a
```

```python
result = {}
yearlist = []
titlelist = []
dbo = []
ibo = []
wbo = []
result['Release year'] = yearlist
result['Title'] = titlelist
result['Domestic'] = dbo
result['International'] = ibo
result['Worldwide'] = wbo
```

```python
for row in allrow[1:101]:
    el = row.findAll('td')
    dbonumber = el[3].contents[-1].replace(',','').strip('$')
    ibonumber = el[4].contents[-1].replace(',','').strip('$')
    wbonumber = el[5].contents[-1].replace(',','').strip('$')
    yearlist.append(el[1].contents[-1])
    titlelist.append(el[2].contents[-1].contents[-1].contents[-1])
    dbo.append(dbonumber)
    ibo.append(ibonumber)
    wbo.append(wbonumber)
print(result)
    
```

```python
df = pd.DataFrame(result)
df.head(10)
```

```python
df = df[['Release year', 'Title', 'Domestic', 'International', 'Worldwide']]
df
```

```python
def get_adjusted_price(year, price):
    webstring = 'https://www.statbureau.org/calculate-inflation-price-json?country=united-states&start=%s/1/1&end=2016/9/1&amount=%s'%(str(year),str(price))
    adjusted = requests.get(webstring)
    if adjusted.status_code == 200:
        adjusted_price = float(json.loads(adjusted.text.replace('$','').replace(' ','')))
    else:
        adjusted_price = math.nan
    return adjusted_price
```

```python
def wrapper(the_row):
    return get_adjusted_price(the_row['Release year'], the_row['Domestic'])
df['adj_domestic']=df.apply(wrapper, axis=1)
```

```python
df.sort_values(by=['adj_domestic'], ascending=False)
```

```python
def getmovieinfo(title, year):
    webstring = 'http://www.omdbapi.com/?t=%s&y=%s&plot=short&r=json'%(str(title),str(year))
    movie = requests.get(webstring)
    movieinfo = (json.loads(movie.text))
    return movieinfo
```

```python
def actors(row):
    info = getmovieinfo(row['Title'], row['Release year'])
    if info['Response'] == 'False':
        return 'Error'
    else:
        actors = info['Actors']
        return actors
df['Actors']=df.apply(actors, axis=1)
```

```python
def genre(row):
    info = getmovieinfo(row['Title'], row['Release year'])
    if info['Response'] == 'False':
        return 'Error'
    else:
        genre = info['Genre']
        return genre
df['Genre']=df.apply(genre, axis=1)
```

```python
def director(row):
    info = getmovieinfo(row['Title'], row['Release year'])
    if info['Response'] == 'False':
        return 'Error'
    else:
        director = info['Director']
        return director
df['Director']=df.apply(director, axis=1)
```
